```python
# Identify predictors and response
train = h2o.H2OFrame(train)
test = h2o.H2OFrame(test)

```

```python
# Identify predictors and response
x = train.columns
y = "Exited"
x.remove(y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('Exited', axis=1), df['Exited'], test_size=0.2, random_state=42)
```

```python
# Define the GBM classifier
gbm = GradientBoostingClassifier()
```

```python
# Define the parameter grid for grid search
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of boosting stages
    'learning_rate': [0.1, 0.05, 0.01],  # Learning rate
    'max_depth': [3, 4, 5]  # Maximum depth of each tree
}
```

```python
# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=gbm, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters and the corresponding model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

```

```python
# Evaluate the model on the test set
accuracy_test = best_model.score(X_test, y_test)
accuracy_train = best_model.score(X_train, y_train)
print(f"Best parameters: {best_params}")
print(f"Accuracy on train set: {accuracy_train}")
print(f"Accuracy on test set: {accuracy_test}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder,FunctionTransformer
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('D:\\UserData\\z004nfzs\\OneDrive - Siemens AG\\Ambiente de Trabalho\\Bank Churn\\Bank-Customer-Churn\\Customer-Churn-Records.csv')
columns_to_remove=['RowNumber','CustomerId','Surname']
df=df.drop(columns=columns_to_remove, axis=1)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('Exited', axis=1), df['Exited'], test_size=0.2, random_state=42)
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of boosting stages
    'learning_rate': [0.1, 0.05, 0.01],  # Learning rate
    'max_depth': [3, 4, 5]  # Maximum depth of each tree
}

# Create an instance of GradientBoostingClassifier
gbm = GradientBoostingClassifier()
```

```python
# Define the columns to be scaled
columns_to_scale = ['column1', 'column2', 'column3']

# Define the columns for one-hot encoding and mapping
columns_to_encode = ['Geography', 'Card Type']
columns_to_map = ['Gender']
```

```python
# Create a ColumnTransformer for preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('normalize', MinMaxScaler(), ['CreditScore','Balance','EstimatedSalary','Point Earned']),
        ('onehot', OneHotEncoder(drop='first'), ['Geography', 'Card Type','Gender'])
    ],
    remainder='passthrough'
)
```

```python
# Create a pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),  # Step for preprocessing, encoding, and mapping
    ('grid_search', GridSearchCV(estimator=gbm, param_grid={
    'n_estimators': [100, 200, 300],  # Number of boosting stages
    'learning_rate': [0.1, 0.05, 0.01],  # Learning rate
    'max_depth': [3, 4, 5]  # Maximum depth of each tree
    }, cv=5))  # Step for grid search
])
```

```python
pipeline.fit(X_train, y_train) 
```

```python
# Access the best parameters and best score
best_params = pipeline.named_steps['grid_search'].best_params_
best_score = pipeline.named_steps['grid_search'].best_score_
```

```python
predictions = pipeline.predict(X_test) 
```
